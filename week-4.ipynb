```python
#Key Findings:
"""
Finding 1: Customers Without Contract Renewal Are Significantly More Likely to Churn
Why this is important

Contract renewal is one of the strongest indicators of customer loyalty. Customers who choose not to renew are much more likely to leave the company.
"""
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='ContractRenewal',
    hue='Churn',
    palette='Set2'
)

plt.title('Contract Renewal vs Customer Churn', fontsize=14)
plt.xlabel('Contract Renewal')
plt.ylabel('Number of Customers')
plt.legend(title='Churn')
sns.despine()
plt.tight_layout()
plt.show()
```

```python
"""
Observation:
Customers who did not renew their contracts show a substantially higher churn rate than customers who renewed. 
In contrast, customers with renewed contracts are much more likely to remain with the company.

Business Interpretation:
Contract renewal reflects customer commitment. 
A lack of renewal suggests reduced loyalty and a greater likelihood of switching to competitors.

Business Impact:
Contract expiration represents a critical opportunity for intervention. Losing customers at this stage increases customer acquisition costs and reduces recurring revenue.
"""
```

```python
"""
Finding 2: Frequent Customer Service Calls Strongly Indicate Churn Risk

Customers usually contact customer support because they are experiencing service issues.
Repeated interactions often indicate unresolved problems.
"""
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x='Churn',
    y='CustServCalls',
    hue='Churn',
    palette='Set2'
)

plt.title('Customer Service Calls vs Churn', fontsize=14)
plt.xlabel('Churn')
plt.ylabel('Customer Service Calls')
sns.despine()
plt.tight_layout()
plt.show()
```

```python

"""
Observation:

Customers who churn generally make more customer service calls than those who remain.
The distribution shows a noticeably higher median number of service calls among churned customers.

Business Interpretation:

Multiple service calls often reflect unresolved complaints or dissatisfaction. 
If issues persist, customers are more likely to cancel their service.

Business Impact:

Customer service interactions can serve as an early warning system for churn. Monitoring these customers enables proactive retention efforts before they leave.
"""
```

```python
"""
Finding 3: Higher Monthly Charges Are Associated With Greater Customer Churn
Pricing directly affects customer satisfaction and perceived value.
"""
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x='Churn',
    y='MonthlyCharge',
    hue='Churn',
    palette='Set2'
)

plt.title('Monthly Charges vs Customer Churn', fontsize=14)
plt.xlabel('Churn')
plt.ylabel('Monthly Charge ($)')
sns.despine()
plt.tight_layout()
plt.show()
```

```python
"""
Observation:
Customers who churn tend to have higher monthly charges than customers who stay. 
The distribution indicates that customers paying premium prices are more likely to discontinue their service.

Business Interpretation:
Higher costs increase customer expectations. 
If customers do not perceive sufficient value, they are more inclined to explore competing providers.

Business Impact:
High-paying customers contribute significantly to revenue.
Their loss has a greater financial impact than losing lower-paying customers.
"""
```

```python
"""
Finding 4: Newer Customers Are More Likely to Leave

The beginning of the customer lifecycle is often the most vulnerable period.
"""
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x='Churn',
    y='AccountWeeks',
    hue='Churn',
    palette='Set2'
)

plt.title('Customer Tenure vs Churn', fontsize=14)
plt.xlabel('Churn')
plt.ylabel('Account Weeks')
sns.despine()
plt.tight_layout()
plt.show()
```

```python
"""
Observation:
Customers who churn generally have fewer account weeks than long-term customers. 
Churn is concentrated among relatively new customers.

Business Interpretation:
Customers who have not yet developed loyalty are more likely to leave if they encounter poor service, pricing concerns, or unmet expectations.

Business Impact:
Retaining customers during their first several months can substantially improve long-term customer lifetime value and reduce acquisition costs.
"""
```

```python
"""
Executive Summary of Key Findings:
Customer retention is a key driver of profitability in the telecommunications industry, making it essential to understand the factors that influence customer churn. 
This analysis examined customer information, including contract renewal status, account tenure, customer service interactions, and monthly charges, to identify the primary reasons customers leave the company.

After cleaning the dataset and performing exploratory data analysis, four major factors were found to be strongly associated with customer churn.
Customers who did not renew their contracts were significantly more likely to leave than those who renewed. 
Customers with a higher number of customer service calls also exhibited a greater likelihood of churning, suggesting that unresolved service issues contribute to customer dissatisfaction. 
In addition, customers with higher monthly charges were more likely to discontinue their service, indicating that pricing may influence retention. 
Finally, customers with shorter account tenure showed higher churn rates, highlighting the importance of engaging new customers during the early stages of their relationship with the company.

These findings indicate that customer churn is driven by a combination of contract decisions, service quality, pricing, and customer loyalty. 
By focusing on these areas, the company can improve customer satisfaction, increase retention, and reduce the costs associated with acquiring new customers.
"""
```

```python
"""
Recommendations 
Recommendation 1: Strengthen Contract Renewal Programs
Customers who do not renew their contracts are at the highest risk of churn. 
The company should introduce personalized renewal campaigns, loyalty discounts, and exclusive offers before contract expiration. 
Early engagement with customers approaching the end of their contracts can improve renewal rates and reduce customer loss.

Recommendation 2: Proactively Support Customers with Frequent Service Calls
Customers making multiple customer service calls often experience unresolved issues and are more likely to churn. 
The company should identify these customers through a monitoring system and provide faster issue resolution, priority customer support, and follow-up communication. 
Addressing customer concerns promptly can significantly improve satisfaction and retention.

Recommendation 3: Improve Value for High-Paying Customers
Customers paying higher monthly charges are more likely to leave if they believe they are not receiving sufficient value. 
The company should review pricing strategies, offer personalized discounts or bundled service packages, and provide additional benefits to premium customers. 
Enhancing perceived value can reduce churn among high-revenue customer segments.

Recommendation 4: Focus on Retaining New Customers
Most churn occurs during the early stages of the customer lifecycle. 
The company should implement a structured onboarding program that includes welcome communications, educational resources, regular follow-ups, and introductory offers during the first few months. 
Building a positive customer experience early can increase long-term loyalty and customer lifetime value.
"""
```
